```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import csv
from collections import Counter
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
with open('data.csv') as csv_file:
    
    csv_reader = csv.DictReader(csv_file)
    
    language_counter = Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguageHaveWorkedWith'].split(';'))
```

```python
languages = []
popularity = []
```

```python
for item in language_counter.most_common(5):
    languages.append(item[0])
    popularity.append(item[1])
```

```python
languages
```

```python
plt.style.use('fivethirtyeight')

plt.pie(popularity, labels=languages)

plt.title('Popularity of Languages')
plt.tight_layout()
```

```python
# adding border

plt.style.use('fivethirtyeight')

plt.pie(popularity, labels=languages, wedgeprops={'edgecolor': 'black'})

plt.title('Popularity of Languages')
plt.tight_layout()
```

```python
# adding custom colors

plt.style.use('fivethirtyeight')

color = ['blue','green','yellow','pink', 'orange']

plt.pie(popularity, labels=languages, colors=color,wedgeprops={'edgecolor': 'black'})

plt.title('Popularity of Languages')
plt.tight_layout()
```

```python
# exploting specific slice

plt.style.use('fivethirtyeight')

explode = [0,0,0.1,0,0]

plt.pie(popularity, labels=languages, explode = explode ,wedgeprops={'edgecolor': 'black'})

plt.title('Popularity of Languages')
plt.tight_layout()
```

```python
# adding shadow

plt.style.use('fivethirtyeight')

explode = [0,0,0.1,0,0]

plt.pie(popularity, labels=languages, explode = explode ,wedgeprops={'edgecolor': 'black'}, shadow=True)

plt.title('Popularity of Languages')
plt.tight_layout()
```

```python
# start angle
  
plt.style.use('fivethirtyeight')

explode = [0,0,0.1,0,0]

plt.pie(popularity, labels=languages, explode = explode ,wedgeprops={'edgecolor': 'black'}, startangle=90)

plt.title('Popularity of Languages')
plt.tight_layout()
```

```python
# showing percentage

plt.style.use('fivethirtyeight')

explode = [0,0,0.1,0,0]

plt.pie(popularity, labels=languages, explode = explode ,wedgeprops={'edgecolor': 'black'}, startangle=90, shadow=True, autopct='%1.1f%%')

plt.title('Popularity of Languages')
plt.tight_layout()
```
