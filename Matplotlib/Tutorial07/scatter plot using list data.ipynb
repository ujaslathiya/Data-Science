```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

```python
plt.style.use('seaborn')
plt.scatter(x,y)
```

```python
#  setting size of the markers

plt.scatter(x,y, s=100)
```

```python
plt.scatter(x,y, s=100, color='green', marker='x')
```

```python
#  adding border and color and border width

plt.scatter(x,y, edgecolor='black', alpha=0.75, linewidth=1,color='green')
```

```python
# adding custom colors

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
plt.scatter(x,y, edgecolor='black', alpha=0.75, linewidth=1, c=colors, cmap='Greens')
```

```python
# adding color bar

plt.scatter(x,y, edgecolor='black', alpha=0.75, linewidth=1, c=colors, cmap='Greens')
cbar = plt.colorbar()
cbar.set_label('Satisfaction')
```

```python
# setting different sizes of the dots

sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x,y, s=sizes,edgecolor='black', alpha=0.75, linewidth=1, c=colors, cmap='Greens')
```

```python
plt.scatter(x,y, s=sizes,edgecolor='black', alpha=0.75, linewidth=1, c=colors, cmap='Greens')
cbar = plt.colorbar()
cbar.set_label('Satisfaction')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/07-ScatterPlots/2019-05-31-data.csv')

df.head(3)
```

```python
views = df['view_count']
likes = df['likes']
ratio = df['ratio']
```

```python
plt.scatter(views,likes,edgecolor='black', alpha=0.75, linewidth=1)
```

```python
plt.scatter(views,likes,edgecolor='black', alpha=0.75, linewidth=1)
plt.xscale('log')
plt.yscale('log')
```

```python
plt.scatter(views,likes,c=ratio,cmap='summer',edgecolor='black', alpha=0.75, linewidth=1)
plt.xscale('log')
plt.yscale('log')
```

```python
plt.scatter(views,likes,c=ratio,cmap='summer',edgecolor='black', alpha=0.75, linewidth=1)
plt.xscale('log')
plt.yscale('log')
cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')
```
