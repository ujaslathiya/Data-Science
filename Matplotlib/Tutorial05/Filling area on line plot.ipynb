```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/05-Fill_Betweens/data.csv')
```

```python
df.head(3)
```

```python
ages = df['Age']
dev_salaries = df['All_Devs']
py_salaries = df['Python']
js_salaries = df['JavaScript']
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(ages, dev_salaries, linestyle='--', color='#444444', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
```

```python
# using fill between method

plt.plot(ages, dev_salaries, linestyle='--', color='#444444', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages, py_salaries, alpha=0.25)

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
```

```python
# filling till median

plt.plot(ages, dev_salaries, linestyle='--', color='#444444', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

median = 57000

plt.fill_between(ages, py_salaries, median, alpha=0.25)

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
```

```python
# fill through where condition

plt.plot(ages, dev_salaries, linestyle='--', color='#444444', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

median = 67000

plt.fill_between(ages, py_salaries, median,where=(py_salaries>median), interpolate = True, alpha=0.25)

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
```

```python
plt.plot(ages, dev_salaries, linestyle='--', color='#444444', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

median = 67000

plt.fill_between(ages, py_salaries, median,where=(py_salaries>median), interpolate = True, alpha=0.25)
plt.fill_between(ages, py_salaries, median,where=(py_salaries<=median), color='red', interpolate = True, alpha=0.25)

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
```

```python
plt.plot(ages, dev_salaries, linestyle='--', color='#444444', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

median = 67000

plt.fill_between(ages, py_salaries, dev_salaries,where=(py_salaries>dev_salaries), interpolate = True, alpha=0.25, label='Above Avg')
plt.fill_between(ages, py_salaries, dev_salaries,where=(py_salaries<=dev_salaries), color='red', interpolate = True, alpha=0.25, label='Below avg')

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
```
