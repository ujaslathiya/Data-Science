```python
import pandas as pd
```

```python
nyc = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/FeatureEngineering/1_outliers/Exercise/AB_NYC_2019.csv')
```

```python
min_thresold, max_thresold = nyc.price.quantile([0.01,0.99])
min_thresold, max_thresold
```

```python
nyc[nyc.price>max_thresold]
```

```python
nyc[nyc.price<min_thresold]
```

```python
nyc2 = nyc[(nyc.price>min_thresold) & (nyc.price<max_thresold)]
nyc2.shape
```
